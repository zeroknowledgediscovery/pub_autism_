```python
import pandas as pd
import pylab as plt
plt.style.use('fivethirtyeight')
%matplotlib inline
font = {'family' : 'Gil Sans',
        'weight' : 'normal',
        'size'   : 18}

plt.rc('font', **font)
```

```python
tf=pd.read_csv('run.dat',sep=' ',header=None)
```

```python
tf=tf.dropna(axis=1)
tf.columns=['patientid','weeks','risk','relative risk','confidence']
```

```python
ax=tf.set_index('weeks')['relative risk'].plot(style='-o',ms=15)
ax.set_ylabel('relative risk')
plt.savefig('../../../../../ehrzero/risk.png',dpi=50,bbox_inches='tight',pad=0,transparent=True)
```
