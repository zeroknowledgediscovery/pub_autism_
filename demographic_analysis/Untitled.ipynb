```python
import pandas as pd
import seaborn as sns
import pylab as plt
plt.style.use('fivethirtyeight')
def saveFIG(filename='tmp.pdf',AXIS=False):
    '''
        save fig for publication
    '''
    import pylab as plt
    plt.subplots_adjust(top = 1, bottom = 0, right = 1, left = 0, 
            hspace = 0, wspace = 0)
    plt.margins(0,0)
    if not AXIS:
        plt.gca().xaxis.set_major_locator(plt.NullLocator())
        plt.gca().yaxis.set_major_locator(plt.NullLocator())
    plt.savefig(filename,dpi=300, bbox_inches = 'tight',
                pad_inches = 0,transparent=True) 
    return
FS=22
plt.rcParams['font.size'] = FS
```

```python
pd.read_csv('UCM_AUC_BY_ETHNICITY.csv').ethnicity.value_counts().index
```

```python
df=pd.read_csv('UCM_AUC_BY_RACE.csv')
COL={'Black/African-American':'r','White':'g','Hispanic or Latino':'b',
     'More than one Race':'darkorange','Asian/Mideast Indian':'m','American Indian or Alaska Native':'b'}
```

```python
dfm=df[df.gender=='M'].set_index('week')
for race in ['Black/African-American','White','More than one Race']:
    plt.fill_between(dfm[dfm.race==race].index, dfm[dfm.race==race].auc+dfm[dfm.race==race].CF, dfm[dfm.race==race].auc-dfm[dfm.race==race].CF,facecolor=COL[race], alpha=0.2)
    plt.plot(dfm[dfm.race==race].index, dfm[dfm.race==race].auc,color=COL[race], lw=3,label=race)
plt.gca().set_ylabel('AUC')
plt.gca().set_xlabel('age [week]')
plt.gca().set_ylim(None,1)
plt.gca().legend(['Black/African-American','White','More than one Race'],fontsize=FS,bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

saveFIG('Race_M.pdf',AXIS=True)
```

```python
dfm=df[df.gender=='F'].set_index('week')
for race in ['Black/African-American','White','More than one Race']:
    plt.fill_between(dfm[dfm.race==race].index, dfm[dfm.race==race].auc+dfm[dfm.race==race].CF, dfm[dfm.race==race].auc-dfm[dfm.race==race].CF,facecolor=COL[race], alpha=0.2)
    plt.plot(dfm[dfm.race==race].index, dfm[dfm.race==race].auc,color=COL[race], lw=3)
plt.gca().set_ylabel('AUC')
plt.gca().set_xlabel('age [week]')
plt.gca().set_ylim(None,1)
saveFIG('Race_F.pdf',AXIS=True)    
```

```python
df=pd.read_csv('UCM_AUC_BY_ETHNICITY.csv')
COL={'Not Hispanic or Latino':'k',
       'Hispanic or Latino':'r'}
```

```python
dfm=df[df.gender=='M'].set_index('week')
for race in ['Not Hispanic or Latino',
       'Hispanic or Latino']:
    plt.fill_between(dfm[dfm.ethnicity==race].index, dfm[dfm.ethnicity==race].auc+dfm[dfm.ethnicity==race].CF, dfm[dfm.ethnicity==race].auc-dfm[dfm.ethnicity==race].CF,facecolor=COL[race], alpha=0.2)
    plt.plot(dfm[dfm.ethnicity==race].index, dfm[dfm.ethnicity==race].auc,color=COL[race], lw=3,label=race)
plt.gca().set_ylabel('AUC')
plt.gca().set_xlabel('age [week]')
plt.gca().set_ylim(None,1)
plt.gca().legend(['Not Hispanic or Latino',
       'Hispanic or Latino'],fontsize=FS,bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

saveFIG('Ethnicity_M.pdf',AXIS=True)
```
